<h3>Average Data to fill in entries that has 0 value</h3>

```python
# Calculate the Average Values
ave_bloodpressure = data['blood_pressure'].mean()
ave_glu_cont = data['glucose_concentration'].mean()
ave_triceps = data['triceps_skin_thickness'].mean()
ave_body_mass_index = data['body_mass_index'].mean()
print(ave_bloodpressure)
print(ave_glu_cont)
print(ave_triceps)
print(ave_body_mass_index)
```

```python
data['body_mass_index']=data['body_mass_index'].replace(0,ave_body_mass_index)
data['triceps_skin_thickness']=data['triceps_skin_thickness'].replace(0,ave_triceps)
data['glucose_concentration']=data['glucose_concentration'].replace(0,ave_glu_cont)
data['blood_pressure']=data['blood_pressure'].replace(0,ave_bloodpressure)
```

<h2>Normalisasi Data</h2>

```python
label = data.pop('is_diabetes')
label.sample(5)
```

```python
# Melakukan normalisasi dengan menggunakan simple feature scaling
for column in list(data.columns):
  data[column] = data[column]/data[column].max()
```

```python
data.sample(5)
```

```python
# Melihat rentang data utk mengetahui apabila scaling berhasil atau tidak, nilai min dan max harus diantara 0 atau 1
for column in list(data.columns):
  print(f"Max value of column {column} is: {round(data[column].max(),3)}")
  print(f"Min value of column {column} is: {round(data[column].min(),3)}")
```

<h2>Begining Our Machine Learning Logistic Regression</h2>

Changing Our data into an array

```python
feature_set = data.to_numpy()
label_set = label.to_numpy()
```

<h3>Building The Model Using Keras</h3>

```python
# importing Tensorflow keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# We are going to use Sequential to make the model, Sequential is a deep learning Model
# Then we add neurons (layers) to the Sequential Model

model = Sequential()
# we are using activation = 'sigmoid' because it is a classification problem, sigmoid will plot our data to max value of 1 and min value of 0
# we also add 1 neuron layer to our model with all our features as the input
model.add(Dense(1, input_dim = len(feature_set[0, :]), activation = 'sigmoid'))
```

```python
# We also tell the computer that it needs to watch out for binary errors because we are doing a logistic regression with 0 and 1 as an output
model.compile(loss = 'binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
# rmsprop is an optimizer algorithm for Artificial Neural Network
# accuracy is commonly used as the metric to evaluate our model
# all of this is pretty much common Rules of Engangement
```

```python
# Now we are going to fit our model, we are going to start training the model
# The verbose argument essential mean that we want to see our machine's learning progress as a progress bar
model.fit(x = feature_set, y= label_set, epochs =512, verbose = 1)
```

```python
# now we test it
# because we did not do a split train and test, we can test it to our training data set (oh well)

# The data that we are going to see
label_set[0:5]
```

```python
predictions = model.predict(feature_set[0:5])
```
